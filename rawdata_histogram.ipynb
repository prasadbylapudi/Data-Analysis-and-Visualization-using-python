```
import matplotlib.pyplot as plt
import numpy as np

# Use numpy to generate a bunch of random data in a bell curve around 5.
n =10+ np.random.randn(1000)#random distribution formulae
#1000 indicates the no of values to be produce to plot
# we can have distribution of Gaussian distribution which is Normal distribution
# function is np.random.normal(paramenters )
#

m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Raw Data")
plt.show()

plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```
